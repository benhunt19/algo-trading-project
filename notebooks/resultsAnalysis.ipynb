```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
sns.set_theme(style="darkgrid")
```

```python
RESULTS_FOLDER = '../data/results'
```

```python
garch_df = pd.read_csv(f'{RESULTS_FOLDER}/portfolio_garch.csv')
gam_df = pd.read_csv(f'{RESULTS_FOLDER}/portfolio_gam.csv')
buy_n_hold_df = pd.read_csv(f'{RESULTS_FOLDER}/portfolio_buy_n_hold.csv')
dfs = [garch_df, gam_df, buy_n_hold_df]
```

```python
actual = pd.read_csv(f'../data/SPTL_2023.csv')
```

```python
# 1. Delta Vt
col = 'PnL'

fig, axes = plt.subplots(3, 1, figsize=(15, 10))
titles = ['GARCH','GAM', 'Buy & Hold', ]
colors = ['blue', 'green', 'orange']

dates = pd.to_datetime(actual['date_string'])

for i, (df, title) in enumerate(zip(dfs, titles)):
    deltaVt = df[col]
    deltaVt.dropna()
    # Set monthly ticks on x-axis
    axes[i].xaxis.set_major_locator(mdates.MonthLocator(interval=1))
    axes[i].xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    plt.setp(axes[i].xaxis.get_majorticklabels(), rotation=30)
    axes[i].plot(dates, deltaVt, color=colors[i])
    mean_deltaVt = deltaVt.mean()
    axes[i].axhline(y=mean_deltaVt, color='r', linestyle='--')
    axes[i].set_title(f'$\Delta V_t$ - {title}')
    axes[i].legend(['$\Delta V_t$', f'Mean = {int(mean_deltaVt)}'])
    axes[i].set_ylabel('$\Delta V_t$ (\$)')
    axes[i].set_xlabel('Date')
    
plt.tight_layout()
```

```python
# 2. Delta Vt cap
col = 'capitalGains'
dfs_no_bnh = dfs[:2]

fig, axes = plt.subplots(2, 1, figsize=(15, 7))
titles = ['GARCH','GAM']
colors = ['blue', 'green']

dates = pd.to_datetime(actual['date_string'])

for i, (df, title) in enumerate(zip(dfs_no_bnh, titles)):
    deltaVtCap = df[col]
    # Set monthly ticks on x-axis
    axes[i].xaxis.set_major_locator(mdates.MonthLocator(interval=1))
    axes[i].xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    plt.setp(axes[i].xaxis.get_majorticklabels(), rotation=30)
    axes[i].plot(dates, deltaVtCap, color=colors[i])
    mean_deltaVtCap = deltaVtCap.mean()
    axes[i].axhline(y=mean_deltaVtCap, color='r', linestyle='--')
    axes[i].set_title('$\Delta V^{cap}_t$ ' + title)
    axes[i].legend(['$\Delta V^{cap}_t$', f'Mean = {int(mean_deltaVtCap)}'])
    axes[i].set_ylabel('$\Delta V^{cap}_t$ (\$)')
    axes[i].set_xlabel('Date')    
    plt.tight_layout(pad=2.0)
```

```python
# 3. Delta Vt_total
col = 'value'

fig, axes = plt.subplots(3, 1, figsize=(15, 10))
titles = ['GARCH','GAM', 'Buy & Hold', ]
colors = ['blue', 'green', 'orange']

dates = pd.to_datetime(actual['date_string'])

for i, (df, title) in enumerate(zip(dfs, titles)):
    deltaVt = df[col] - df[col].shift(1)
    deltaVt.dropna()
    # Set monthly ticks on x-axis
    axes[i].xaxis.set_major_locator(mdates.MonthLocator(interval=1))
    axes[i].xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    plt.setp(axes[i].xaxis.get_majorticklabels(), rotation=30)
    axes[i].plot(dates, deltaVt, color=colors[i])
    mean_deltaVt = deltaVt.mean()
    axes[i].axhline(y=mean_deltaVt, color='r', linestyle='--')
    axes[i].set_title('$\Delta V^{total}_t$ ' + title)
    axes[i].legend(['$\Delta V^{total}_t$', f'Mean = {int(mean_deltaVt)}'])
    axes[i].set_ylabel('$\Delta V^{total}_t$ (\$)')
    axes[i].set_xlabel('Date')
    
plt.tight_layout()
```

```python
# Cumulative Delta V
col = 'PnL'

fig, ax = plt.subplots(figsize=(15, 7))
titles = ['GARCH','GAM', 'Buy & Hold']
colors = ['blue', 'green', 'orange']

dates = pd.to_datetime(actual['date_string'])

for i, (df, title) in enumerate(zip(dfs, titles)):
    deltaVt = df[col]
    deltaVtCS = deltaVt.cumsum()
    deltaVtCS.dropna()
    ax.plot(dates, deltaVtCS, color=colors[i], label=title)
    mean_deltaVtCS = deltaVtCS.mean()
    ax.axhline(y=mean_deltaVtCS, color=colors[i], linestyle='--', alpha=0.5, label=f'{title} Mean Position = ${int(mean_deltaVtCS)}')
    # Add fill color based on values
    ax.fill_between(dates, deltaVtCS, 0, where=(deltaVtCS >= 0), color='green', alpha=0.1)
    ax.fill_between(dates, deltaVtCS, 0, where=(deltaVtCS < 0), color='red', alpha=0.1)

# Set monthly ticks on x-axis
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))
ax.yaxis.set_major_locator(plt.MaxNLocator(20))  # Increase number of y-axis ticks
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.setp(ax.xaxis.get_majorticklabels(), rotation=30)
ax.axhline(y=0, color='black', linestyle='-', alpha=0.3, label='Zero Returns')
ax.set_title('$V_t$ Over Full Time Horizon')
ax.legend()
ax.set_ylabel('$V_t$ (\$)')
ax.set_xlabel('Date')
    
plt.tight_layout()
```

```python
# Cumulative Delta V Cap
col = 'value'
starting = 1_000_000

fig, ax = plt.subplots(figsize=(15, 7))
titles = ['GARCH','GAM', 'Buy & Hold']
colors = ['blue', 'green', 'orange']

dates = pd.to_datetime(actual['date_string'])

for i, (df, title) in enumerate(zip(dfs, titles)):
    totalVal = df[col] - starting
    ax.plot(dates, totalVal, color=colors[i], label=title)
    mean_totalVal = totalVal.mean()
    ax.axhline(y=mean_totalVal, color=colors[i], linestyle='--', alpha=0.5, label=f'{title} Mean Position = ${int(mean_totalVal)}')
    # Add fill color based on values
    ax.fill_between(dates, totalVal, 0, where=(totalVal >= 0), color='green', alpha=0.1)
    ax.fill_between(dates, totalVal, 0, where=(totalVal < 0), color='red', alpha=0.1)

# Set monthly ticks on x-axis
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))
ax.yaxis.set_major_locator(plt.MaxNLocator(20))  # Increase number of y-axis ticks
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.setp(ax.xaxis.get_majorticklabels(), rotation=30)
ax.axhline(y=0, color='black', linestyle='-', alpha=0.3, label='Zero Returns')
ax.set_title('$V^{total}_t$ Over Full Time Horizon')
ax.legend()
ax.set_ylabel('$V^{total}_t$ (\$)')
ax.set_xlabel('Date')
    
plt.tight_layout()
```

```python
# Metrics analysis
# Create the metrics DataFrame with proper data types
metrics = pd.DataFrame([
    ["Profit/Loss ($)", 50727.71, 12198.28, -45506.08],
    ["% Increase", 50.73, 12.20, -45.51],
    ["Sharpe Ratio", 0.4749, 0.1598, -0.1979],
    ["Max. Drawdown", 0.1454, 0.2313, 0.11494], 
    ["Calmar Ratio", 0.3516, 0.1070, -0.1983]
], columns=["Metric", "GARCH", "GAM", "Buy and Hold"])

# Create a separate DataFrame for normalized values, excluding the Metric column
metrics_norm = metrics.iloc[:, 1:].copy()
# Normalize each metric
metrics_norm.iloc[0] = (metrics_norm.iloc[0] + 50000) / 100000  # Normalize Profit/Loss
metrics_norm.iloc[1] = (metrics_norm.iloc[1] + 50) / 100        # Normalize % Increase
metrics_norm.iloc[2] = (metrics_norm.iloc[2] + 0.5) / 1         # Normalize Sharpe Ratio
metrics_norm.iloc[3] = metrics_norm.iloc[3] / 0.25              # Normalize Max Drawdown
metrics_norm.iloc[4] = (metrics_norm.iloc[4] + 0.5) / 1         # Normalize Calmar Ratio

# Set up the angles
labels = metrics['Metric'].tolist()
num_vars = len(labels)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# Make the plot circular
angles += angles[:1]

# Create figure and polar axis
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(projection='polar'))

# Plot data
for strategy, color in zip(['GARCH', 'GAM', 'Buy and Hold'], ['blue', 'green', 'orange']):
    values = metrics_norm[strategy].tolist()
    values += values[:1]
    ax.plot(angles, values, color=color, linewidth=1, label=strategy)
    ax.fill(angles, values, color=color, alpha=0.25)

# Fix axis to go in the right order and start at 12 o'clock
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)

# Draw axis lines for each angle and label
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)
# Set the legend to top right corner
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
# Add legend
# plt.legend(loc='upper right', bbox_to_anchor=(0.3, 0.1))

plt.title("Normalised Strategy Performance Comparison Metrics", pad=20)
plt.tight_layout()
plt.show()
```
